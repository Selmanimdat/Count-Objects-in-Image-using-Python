```python
import cv2
import matplotlib.pyplot as plt
import cvlib
from cvlib.object_detection import draw_bbox
from collections import Counter
```

```python
# define a function for count objects√ú
```

```python
def count_objects(img_path):
    try:
        #read image
        img=cv2.imread(img_path)
    except:
        print("cant find image")
    #detect objects
    box,label,count=cvlib.detect_common_objects(img)
    output=draw_bbox(img,box,label,count)
    #show image countly
    plt.figure(figsize=(10,10))
    plt.imshow(output);
    #print count of objects
    counts = Counter(label)
    result = '\n'.join(map(lambda item: f"{item[0]}: {item[1]}", counts.items()))
    print(result)
```

```python
#sample using
```

```python
count_objects("cars.jpg")
```
